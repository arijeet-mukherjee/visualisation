<font color='green'>Multivariate Regression is a method used to measure the degree at which more than one independent variable (predictors) and more than one dependent variable (responses), are linearly related. The method is broadly used to predict the behavior of the response variables associated to changes in the predictor variables, once a desired degree of relation has been established.
Linear regression performs the task to predict a dependent variable value (y) based on a given independent variable (x). So, this regression technique finds out a linear relationship between x (input) and y(output). Hence, the name is Linear Regression.

Preapre your x and y values. x inculdes all the features and y the last column or dependent value

```python
X = df.iloc[:, :-1].values 
Y = df.iloc[:, -1].values 
```

```python
X[:,1].shape
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train, X_test, y_train, y_test = np.asarray(train_test_split(X, Y, test_size=5))
```

```python
from sklearn import linear_model
```

```python
model=linear_model.LinearRegression()
```

```python
model.fit(X_train,y_train)
```

```python
prediction=model.predict(X_test)
```

```python
from sklearn.metrics import r2_score
```

```python
r2_score(y_test,prediction)
```
